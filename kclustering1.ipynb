```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('income.csv')
```

```python
df1=df.copy()
```

```python
df1.head()
```

```python
df1.drop('Name',axis=1,inplace=True)
```

```python
df1
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('age')
plt.ylabel('Income($)')
```

```python
#by seeing scatter plot we can clearly see there are 3 group of clusters
km=KMeans(n_clusters=3,init='k-means++')
y_label=km.fit_predict(df1)
```

```python
df1['y_label']=y_label
```

```python
df1
```

```python
#creating 3 dataframes for each clusters
dfa=df1[df1['y_label']==0]
dfb=df1[df1['y_label']==1]
dfc=df1[df1['y_label']==2]
```

```python
plt.scatter(dfa.Age,dfa['Income($)'],c='red',s=60,label='cluster1')
plt.scatter(dfb.Age,dfb['Income($)'],c='green',s=60,label='cluster2')
plt.scatter(dfc.Age,dfc['Income($)'],c='purple',s=60,label='cluster3')
plt.xlabel('AGE')
plt.ylabel('Income($')
plt.legend()
plt.show()
```

```python
#by seeing above scatter plot we can conclude some of the data points are misclassified .
#Since age and income features are in different scales that leads to the misclassification.so we will apply scaling on these  columns
scaler=MinMaxScaler()

df['Age']=scaler.fit_transform(df[['Age']])
df['Income($)']=scaler.fit_transform(df[['Income($)']])
```

```python
df
```

```python
df2=df.copy()
```

```python
df2.drop('Name',axis=1,inplace=True)
```

```python
df2
```

```python
km=KMeans(n_clusters=3,init='k-means++')
y_pred2=km.fit_predict(df2)
```

```python
df2['y_pred2']=y_pred2
```

```python
df2
```

```python
#creating 3 dataframes for each clusters
dfaa=df2[df2['y_pred2']==0]
dfbb=df2[df2['y_pred2']==1]
dfcc=df2[df2['y_pred2']==2]
```

```python
km.cluster_centers_
```

```python
plt.scatter(dfaa.Age,dfaa['Income($)'],c='red',s=60,label='cluster1')
plt.scatter(dfbb.Age,dfbb['Income($)'],c='green',s=60,label='cluster2')
plt.scatter(dfcc.Age,dfcc['Income($)'],c='purple',s=60,label='cluster3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='black',s=190,marker='*',label='centroid')
plt.xlabel('AGE')
plt.ylabel('Income($')
plt.legend()
plt.show()
```

```python
#elobow curve
wcss=[]
k=range(3,11)

for i in k:
  km=KMeans(n_clusters=i,init='k-means++').fit(df2[['Age','Income($)']])
  wcss.append([i,km.inertia_])
```

```python
wcss
wcss=pd.DataFrame(wcss)
```

```python
plt.plot(wcss[0],wcss[1])
plt.xlabel('k')
plt.ylabel('wcss')
```

```python
#silhoutee score

from sklearn.metrics import silhouette_score

#elobow curve
sil=[]
k=range(3,11)

for i in k:
  km=KMeans(n_clusters=i,init='k-means++').fit(df2[['Age','Income($)']])
  sil.append([i,silhouette_score(df2.iloc[:,:1],km.labels_)])
```

```python
sil=pd.DataFrame(sil)
```

```python
plt.plot(sil[0],sil[1])
plt.xlabel('k value')
plt.ylabel('silhoutee score')
```
